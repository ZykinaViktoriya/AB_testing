```python
import numpy as np
```

```python
groupped_df = df.copy()
groupped_df['rel_amount'] = groupped_df['amount'] / sum(groupped_df['amount'])
groupped_df = groupped_df.sort_values('rel_amount',ascending=False)
groupped_df['cumsum_amount'] = groupped_df['rel_amount'].cumsum()
groupped_df['abc_amount'] = np.where(
    groupped_df['cumsum_amount']<0.8,
    'A',
      np.where(
          groupped_df['cumsum_amount']<0.95,
               'B',
              'C'))
```

```python
columns = list(df.columns)
columns.remove('name')

groupped_df = df.copy()

for col in columns:
  groupped_df[f'rel_{col}'] = groupped_df[col] / sum(groupped_df[col])
  groupped_df = groupped_df.sort_values(f'rel_{col}',ascending=False)
  groupped_df[f'cumsum_{col}'] = groupped_df[f'rel_{col}'].cumsum()
  groupped_df[f'abc_{col}'] = np.where(
    groupped_df[f'cumsum_{col}']<0.8,
    'A',
      np.where(
          groupped_df[f'cumsum_{col}']<0.95,
               'B',
              'C'))
```

```python
groupped_df = groupped_df[['name', 'revenue', 'amount', 'profit','abc_revenue', 'abc_amount', 'abc_profit']]
```

```python
groupped_df['final_group'] = groupped_df['abc_revenue'] + groupped_df['abc_amount'] + groupped_df['abc_profit']
```

```python
groupped_df
```

```python
groups = groupped_df.groupby('final_group')['final_group'].agg({'count'}).reset_index()
groups
```

```python
import plotly.express as px

fig = px.treemap(groups, path=['final_group'], values = 'count')
fig.show()
```
